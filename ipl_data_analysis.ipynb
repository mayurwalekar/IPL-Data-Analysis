```python
#Combining all files

import pandas as pd
file1="IPL_2008.csv"
file2="IPL_2009.csv"
file3="ipl 2010.csv"
file4="ipl 2011.csv"
file5="ipl 2012.csv"
file6="ipl 2013.csv"
file7="ipl 2014.csv"
file8="ipl 2015.csv"
file9="ipl 2016.csv"
file10="ipl 2016.csv"
file11="ipl 2017.csv"
file12="ipl 2018.csv"
file13="ipl 2019.csv"
file14="ipl 2020.csv"
file15="ipl 2021.csv"
file16="ipl 2022.csv"

#Concatinate
df3 = pd.concat(
   map(pd.read_csv, [file1,file2,file3,file4,file5,file6,file7,file8,file9,file10,file11,file12,file13,file14,file15,file16]), ignore_index=False)
df3.head()
```

```python
df3.to_csv('IPL_Data.csv',index=False)
```

```python
df3.head(5)
```

```python
import numpy as np 
import os
import seaborn as sns
import warnings
import matplotlib.pyplot as plt 
%matplotlib inline
pd.set_option('display.max_columns', None)
warnings.simplefilter("ignore")
sns.set_style('darkgrid')
```

```python
 #barplot of top 5 most player_of_match
pom = df3['Player of the match'].value_counts().nlargest(5)
plt.figure(figsize=(10, 5))
sns.barplot(pom.index, pom)
plt.xlabel('Players Name')
plt.ylabel('Count')
```

```python
# barplot of most toss winners
toss_win_label = df3['Toss Winner'].value_counts()
plt.figure(figsize=(10, 8))
sns.barplot(toss_win_label, toss_win_label.index);
plt.ylabel('Teams')
plt.xlabel('Count')
plt.title('Toss Winners')
```

```python
#teams did batting first and won
win = df3[(df3['Toss Decision'] == 'Bat') & (df3['Won by']== 'Runs')]['Match Winner'].value_counts().sort_values(ascending=False)
plt.figure(figsize=(10, 5))
sns.barplot(win, win.index)
plt.xlabel('Count')
plt.ylabel('Teams')
```

```python
#teams did bowling first and won
win = df3[(df3['Toss Decision'] == 'Field') & (df3['Won by']== 'Wickets')]['Match Winner'].value_counts().sort_values(ascending=False)
plt.figure(figsize=(10, 5))
sns.barplot(win, win.index)
plt.xlabel('Count')
plt.ylabel('Teams')
```

```python
# barplot of most toss winners
toss_win_label = df3['Match Winner'].value_counts()
plt.figure(figsize=(10, 8))
sns.barplot(toss_win_label, toss_win_label.index);
plt.ylabel('Teams')
plt.xlabel('Count')
plt.title('Match Winners')
```

```python
exp_vals = df3['Match Winner'].value_counts()
exp_labels = exp_vals.index
plt.pie(exp_vals,labels=exp_labels,shadow=True, autopct='%0.2f%%',radius=3.5)
plt.show()
```
